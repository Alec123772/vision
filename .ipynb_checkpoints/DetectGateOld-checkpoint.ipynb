```python
def importImage(file):
    img = cv2.imread(file)
    return img

def threshhold(img):
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    #mask = cv2.inRange(hsv, (80,0,0), (95, 255, 255))
    mask = cv2.inRange(hsv, (0,0,0), (255, 255, 255))
    target = cv2.bitwise_and(img,img, mask=mask)
    return target
def binarization(img):
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    ret,thresh1 = cv2.threshold(gray,220,255,cv2.THRESH_BINARY_INV)
    thresh1 = cv2.bitwise_not(thresh1)
    return thresh1

def deNoise(img):
    dst = cv2.fastNlMeansDenoisingColored(img,None,10,0,7,21)
    return dst
    
def incContrast(img):
    # CLAHE (Contrast Limited Adaptive Histogram Equalization)
    clahe = cv2.createCLAHE(clipLimit=3., tileGridSize=(10,10))

    lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)  # convert from BGR to LAB color space
    l, a, b = cv2.split(lab)  # split on 3 different channels

    l2 = clahe.apply(l)  # apply CLAHE to the L-channel

    lab = cv2.merge((l2,a,b))  # merge channels
    dst = cv2.cvtColor(lab, cv2.COLOR_LAB2BGR)  # convert from LAB to BGR
    return dst

def CannyThreshold(img):
    v = np.median(img)
    # apply automatic Canny edge detection using the computed median
    sigma = 0.5
    lower = int(max(0, (1.0 - sigma) * v))
    upper = int(min(255, (1.0 + sigma) * v))
    dst = cv2.Canny(img, lower, upper)
    return dst

def sharpen(img):
    kernel = np.array([[-1,-1,-1], [-1,9,-1], [-1,-1,-1]])
    im = cv2.filter2D(img, -1, kernel)
    return im
    
def filterImage(img):
    dst = cv2.GaussianBlur(img, (0, 0), 3);
    dst = cv2.addWeighted(img, 1.5, dst, -0.5, 0);
    return dst
    
def removeShadow(img):
    rgb_planes = cv2.split(img)
    result_planes = []
    result_norm_planes = []
    for plane in rgb_planes:
        dilated_img = cv2.dilate(plane, np.ones((10,10), np.uint8))
        bg_img = cv2.medianBlur(dilated_img, 43)
        diff_img = 255 - cv2.absdiff(plane, bg_img)
        norm_img = cv2.normalize(diff_img, None, alpha=0, beta=255, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_8UC1)
        result_norm_planes.append(norm_img)
    dst = cv2.merge(result_norm_planes)
    return dst

def detectShape(img):
    (_,contours,_) = cv2.findContours(img, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)
    dst = cv2.cvtColor(img, cv2.COLOR_GRAY2RGB)
    for contour in contours:
        area = cv2.contourArea(contour)
        if area>100 and area<10000:
            print(area)
            rect = cv2.minAreaRect(contour)
            box = cv2.boxPoints(rect)
            box = np.int0(box)
            print(box)
            side1 = math.sqrt((box[0][0]-box[1][0])*(box[0][0]-box[1][0])+(box[0][1]-box[1][1])*(box[0][1]-box[1][1]))
            side2 = math.sqrt((box[2][0]-box[1][0])*(box[2][0]-box[1][0])+(box[2][1]-box[1][1])*(box[2][1]-box[1][1]))
            contArea =side1*side2
            print(side1)
            print(side2)
            print(contArea)
            density = area/contArea
            print(density)
            if True:
                cv2.drawContours(dst,[box],0,(0,0,255),2)
    return dst
    
    
def addLines(img):
    vecSum = cv2.reduce(img, 0, CV_REDUCE_SUM, CV_32S);
    max1=0;
    max2=1;
    for i in range(2,len(vecSum)):
        if vecSum[i] > max2:
            if vecSum[i] > max1:
                max2 = max1
                max1 = i
            else:
                max2 = i
    cv2.line(img, (max1, 0), (max2, y2), (0,255,0), 5)
    return img



def showImage(original, dst):  
    cv2.imshow('Original', original)
    cv2.imshow('New', dst)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
def removeShadow(img):
    rgb_planes = cv2.split(img)
    result_planes = []
    result_norm_planes = []
    for plane in rgb_planes:
        dilated_img = cv2.dilate(plane, np.ones((17,3), np.uint8))
        bg_img = cv2.medianBlur(dilated_img, 21)
        diff_img = 255-cv2.absdiff(plane, bg_img)
        norm_img = cv2.normalize(diff_img, None, alpha=0, beta=255, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_8UC1)
        result_norm_planes.append(norm_img)
    dst = cv2.merge(result_norm_planes)
    return dst
def binary(img):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    return cv2.adaptiveThreshold(~gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 15, -2);

def getLines(newImg):
    csums = np.sum(newImg, axis=0)
    csums1 = copy.deepcopy(csums)
    lineLocs = []
    for i in range(3):
        lineLocs.append([np.argmin(csums),csums[np.argmin(csums)]])
        csums[lineLocs[i][0]-50:lineLocs[i][0]+50]=1000000
    if True:
        plt.plot(csums1)
        for i in range(len(lineLocs)):
            plt.axvline(x=lineLocs[i][0], color='r', linewidth=1)
        plt.show()
    newImg = cv2.cvtColor(newImg, cv2.COLOR_GRAY2BGR)
    return lineLocs
def getLines1(newImg):
    csums = np.sum(newImg, axis=0)
    lineLocs = argrelextrema(csums, np.less)
    if True:
        plt.plot(csums)
        for i in range(len(lineLocs)):
            plt.axvline(x=lineLocs[i][0], color='r', linewidth=1)
        plt.show()
    return lineLocs
    
def erode(img):
    kernel = np.ones((4,1), np.uint8) 
    dst = cv2.erode(img, kernel, iterations=1)
    #kernel = np.ones((4,1), np.uint8) 
    #dst = cv2.dilate(img, kernel, iterations=1)
    return dst

def removeBad(img):
    csums = np.sum(img, axis=0)
    print(csums)
    if True:
        plt.plot(csums)
        plt.show()
        
def plotLines(lineLocs, original):
    for i in range(3):
        cv2.line(original, (lineLocs[i][0], 0), (lineLocs[i][0], original.shape[0]), (0,255,0), 3)
    norm = 0
    center = 0
    for k in range(len(lineLocs)):
        center = center + (50000-lineLocs[k][1])*lineLocs[k][0]
        norm = norm + (50000-lineLocs[k][1])
    center = (int) (center/norm)
    cv2.line(original, (center, 0), (center, original.shape[0]), (0,0,255), 1)
    return original

def segment(image):
    mdpt = (int)(image.shape[0]/2)
    striph = 100
    return image[mdpt - striph: mdpt + striph,:]

def findLines(img):
    gray = cv2.imread('lines.jpg')
    edges = cv2.Canny(gray,50,150,apertureSize = 3)
    cv2.imwrite('edges-50-150.jpg',edges)
    minLineLength=50
    lines = cv2.HoughLinesP(image=edges,rho=1,theta=np.pi/180, threshold=100,lines=np.array([]), minLineLength=minLineLength,maxLineGap=80)
    
    a,b,c = lines.shape
    for i in range(a):
        cv2.line(gray, (lines[i][0][0], lines[i][0][1]), (lines[i][0][2], lines[i][0][3]), (0, 0, 255), 3, cv2.LINE_AA)
        cv2.imwrite('houghlines5.jpg',gray)
    
    
```

```python
def getBoundingBox(img, thresh=.01, buff=0.05, black=False):
    if black:
        thresh = .99
    r = img.shape[0]
    c = img.shape[1]
    
    cmeans = np.mean(img, axis=0)
    cmean_thresh = np.max(cmeans) * thresh
    if black:
        cidxs_above = np.argwhere(cmeans < cmean_thresh)
    else:
        cidxs_above = np.argwhere(cmeans > cmean_thresh)
    col_start = cidxs_above[0][0]
    col_end = cidxs_above[-1][0]
    
    rmeans = np.mean(img, axis=1)
    rmean_thresh = np.max(rmeans) * thresh
    if black:
        ridxs_above = np.argwhere(rmeans < rmean_thresh)
    else:
        ridxs_above = np.argwhere(rmeans > rmean_thresh)
    row_start = ridxs_above[0][0]
    row_end = ridxs_above[-1][0]
    
    # add buffer to ignore clefs
    bufflen = np.int((col_end - col_start) * buff)
    col_start_adj = col_start + bufflen
    
    return (row_start, col_start_adj), col_end - col_start_adj, row_end - row_start
```

```python
def showImageWithBB(X, loc, w, h, sz=(12,12)):
    fig = plt.figure(figsize=sz)
    plt.imshow(1-X, cmap='gray')
    ax = fig.gca()
    rect = patches.Rectangle((loc[1],loc[0]), w, h, linewidth=1,edgecolor='r',facecolor='none')
    ax.add_patch(rect)
    plt.show()  
```

```python
def getCropped(X, black=False):
    ll, w, h = getBoundingBox(X, black=black)
    return X[ll[0]:ll[0]+h,ll[1]:ll[1]+w]
```

```python
def getStripBoundaries(X, showPlot = False, threshold=0.2):
    
    # find staves
    filt = np.vstack((np.ones(21),np.zeros(21),-1*np.ones(21))) # emphasize horizontal lines
    Xfilt = signal.convolve2d(X-np.mean(X), filt, mode='same')
    rsum = np.sum(Xfilt, axis=1)
    rsum_new = []
    for i in range(rsum.shape[0]-4):
        rsum_new.append(sum(rsum[i:i+4]))
    rsum = np.array(rsum_new)
    rsum_spectra = librosa.core.stft(rsum,n_fft=32,hop_length=1) # look for high frequency oscillations
    spectra_max = np.max(rsum_spectra[4:10,:], axis=0)
    spectra_max_smoothed = np.convolve(spectra_max,np.hanning(15),'same')
    peakidxs = signal.find_peaks_cwt(spectra_max_smoothed, range(15,41,1))

    maxval = np.max(np.abs(spectra_max_smoothed))
    peakidxs_threshold = [peakidx for peakidx in peakidxs if 
                          np.abs(spectra_max_smoothed[peakidx])/maxval > threshold]
    peakidxs = peakidxs_threshold
    
    #if len(peakidxs_threshold) == 1 or peakidxs[0] > center or peakidxs[-1] < center:
    #    threshold -= 0.1
    #    peakidxs_threshold = [peakidx for peakidx in peakidxs if 
    #                      np.abs(spectra_max_smoothed[peakidx])/maxval > threshold]

    # determine boundary
    #print(peakidxs)
    boundaries = []
    center = np.int(X.shape[0]/2)
    upper = [p for p in peakidxs if p < center]
    lower = [p for p in peakidxs if p > center]
    
    # show peak picking results (for debugging)
    if showPlot:
        plt.plot(spectra_max_smoothed)
        for pk in peakidxs:
            plt.axvline(x=pk, color='r', linewidth=1)
            
    if upper == [] or lower == []:
        showImage1(X)
        print(peakidxs)
    
    right_peak = upper[-1]
    left_peak = lower[0]
    spectra_mid = spectra_max_smoothed[right_peak:left_peak] 
    mid = np.int(np.argmin(spectra_mid) + right_peak)
    height = np.int((left_peak - right_peak)*1.125)
    #print(right_peak, left_peak, mid, height)
    boundaries = [mid-height, mid, mid+height]
            
    return boundaries, peakidxs
```

```python
def getStrips(X, bnds):
    strips = []
    for i in range(len(bnds)-1):
        rstart = bnds[i]
        rend = bnds[i+1] + 1 # include endpoint
        chunk = X[rstart:rend,:]
        strips.append(chunk)
    return strips
```

```python
def locateStaffLines(s, min_height = 30, max_height = 100, plot = False):
    #print(min_height, max_height)
    s,_ = rotateToHorizontal(s)
    
    L = s.shape[1]
    rsums = np.sum(s, axis=1)
    midpt = np.round(s.shape[0] * .5).astype('int')
    bestScore = 0
    lineLocs = np.zeros(5)
    for h in range(min_height,max_height+1):
        idxs = h * np.arange(5) / 4.0
        idxs = idxs.round().astype('int')
        filt = np.zeros(h+1)
        filt[idxs] = 1 # create comb filter
        scores = np.convolve(rsums, filt, 'valid')
        idx = np.argmax(scores)
        score = scores[idx]
        curScore = score
        if curScore > bestScore:
            bestScore = curScore
            lineLocs = idxs + idx
    
    if plot:
        plt.plot(rsums)
        for i in range(len(lineLocs)):
            plt.axvline(x=lineLocs[i], color='r', linewidth=1)
        plt.show()
        
    return lineLocs
```
